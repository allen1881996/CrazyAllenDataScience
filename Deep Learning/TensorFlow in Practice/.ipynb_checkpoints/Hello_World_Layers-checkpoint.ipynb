```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

In tensorflow, dense is layer, so this one is a simply one layer, single neuron.

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

The code define loss functions and optimizer, the loss function measure how our model guess for a Y value, give the data to optimizer which figure out the next guess. Each guess should be better than one before. When the accruacy is 100%, we can say convergence.

In this case, the loss is mean squared error, and the optimizer is SGD- stochastic gradient descent.

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

Represent the known data.

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

Training, using fit command, epochs means how many times we go through the training loop.

```python
model.fit(xs, ys, epochs=500)
```

```python
print(model.predict([10.0]))
```
