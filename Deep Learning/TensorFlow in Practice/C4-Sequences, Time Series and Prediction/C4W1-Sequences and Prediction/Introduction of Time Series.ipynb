Time Series: It's typically defined as an ordered sequence of values that are usually equally spaced over time.

- univariate
- multivariate

Commen Patterns:
- trend: time series have a specific direction that they're moving in. 
- seasonality: which is seen when patterns repeat at predictable intervals. 
- white noise: that are probably not predictable at all and just a complete set of random values producing.
- auto correlation: Namely it correlates with a delayed copy of itself often called a lag. This example you can see at lag one there's a strong autocorrelation. Often a time series like this is described as having memory as steps are dependent on previous ones. The spikes which are unpredictable are often called Innovations.

- non-stationary time series
- stationary: behavior does not change over time.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot a series
def plot_series(time, series):
    plt.figure(figsize=(10, 6))
    plt.plot(time, series)
    plt.xlabel("time")
    plt.ylabel("value")
    plt.grid(True)
    plt.show()
```

```python
# return a trend 
def trend(time, slope=0):
    return slope * time
```

```python
time = np.arange(4 * 365 + 1)
baseline = 10
series = trend(time, 0.1)
plot_series(time, series)
```

```python
# add a seaonal pattern
def seasonal_pattern(season_time):
    return np.where(season_time < 0.4,
                    np.cos(season_time * 2 * np.pi),
                    1 / np.exp(3 * season_time))

def seasonality(time, period, amplitude=1, phase=0):
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time)
```

```python
baseline = 10
amplitude = 40
series = seasonality(time, period=365, amplitude=amplitude)
plot_series(time, series)
```

```python
slope = 0.05
series = baseline + trend(time, slope) + seasonality(time, period=365, amplitude=amplitude)
plot_series(time, series)
```

```python
def noise(time, noise_level=1):
    return np.random.randn(len(time)) * noise_level
```

```python
noise_level = 15
noisy_series = series + noise(time, noise_level)
plot_series(time, noisy_series)
```

```python
noise_level = 40
noisy_series = series + noise(time, noise_level)
plot_series(time, noisy_series)
```

```python
def autocorrelation(time, amplitude):
    rho1 = 0.5
    rho2 = -0.1
    ar = np.random.randn(len(time) + 50)
    ar[:50] = 100
    for step in range(50, len(time) + 50):
        ar[step] += rho1 * ar[step - 50]
        ar[step] += rho2 * ar[step - 33]
    return ar[50:] * amplitude
```

```python
def autocorrelation(time, amplitude):
    rho = 0.8
    ar = np.random.randn(len(time) + 1)
    for step in range(1, len(time) + 1):
        ar[step] += rho * ar[step - 1]
    return ar[1:] * amplitude
```

```python
series = autocorrelation(time, 10)
plot_series(time[:200], series[:200])
```

```python
series = noise(time)
plot_series(time[:200], series[:200])
```

```python
series = autocorrelation(time, 10) + trend(time, 2)
plot_series(time[:200], series[:200])
```

```python
series = autocorrelation(time, 10) + seasonality(time, period=50, amplitude=150) + trend(time, 2)
plot_series(time[:200], series[:200])
```
