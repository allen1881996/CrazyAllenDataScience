接下来，我们希望做一张图，随着时间的推移，不同地区AveragePrcie的变化

```python
# figure out all of the regions
df.values
```

```python
df['region'].values
```

```python
# convert it to a list, then convert it to a set.
# then convert it to a list again, and we can iterate
list(set(df['region'].values.tolist()))
```

```python
# a simply way
df['region'].unique()
```

```python
# create a dataframe
graph_df = pd.DataFrame()
# iterate over all of the region, 对于每一个region，我们要计算rolling mean of average price
for region in df['region'].unique()[:16]:
    print(region)
    # create region_df, 它是df的复制, where the region is what we are working with
    region_df = df.copy ()[df['region']==region]
    # 比如，当region == Albany, region_df得到
    region_df.set_index('Date', inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f"{region}_price25ma"] = region_df["AveragePrice"].rolling(25).mean()
    # f is used to give us a unique column name each time
    # if this dataframe does not have anything in it
   
    if graph_df.empty:
        graph_df = region_df[[f"{region}_price25ma"]]  
        # note the double square brackets! 因为如果只有一个，就是一个series，两个的话就是一个dateframe，后续我们可以添加其他东西
    else:
        # bring dataframe together on an index
        # you have two dateframe and they are indexed in the exact same way
        # first_df.join.second_df, they will be join on their index
        graph_df = graph_df.join(region_df[f"{region}_price25ma"])
```

```python
df = df.copy()[df['type']=='organic']

df["Date"] = pd.to_datetime(df["Date"])

df.sort_values(by="Date", ascending=True, inplace=True)
df.head()
```

```python
graph_df = pd.DataFrame()
for region in df['region'].unique():
    print(region)
    region_df = df.copy ()[df['region']==region]
    region_df.set_index('Date', inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f"{region}_price25ma"] = region_df["AveragePrice"].rolling(25).mean()
   
    if graph_df.empty:
        graph_df = region_df[[f"{region}_price25ma"]]  
    else:
        graph_df = graph_df.join(region_df[f"{region}_price25ma"])
```

```python
graph_df.dropna().plot(figsize=(15,8), legend=False)
```
