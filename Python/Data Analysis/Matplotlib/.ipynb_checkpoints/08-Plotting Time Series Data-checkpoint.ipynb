```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle = 'solid')
# gcf: got current figure, then autoformat date
plt.gcf().autofmt_xdate() 
date_formate = mpl_dates.DateFormatter('%b/%d/%Y') # format datetime
plt.gca().xaxis.set_major_formatter(date_formate) # set the format to out x-axis, gca is the get current axis

plt.tight_layout() # add a padding to our plot

plt.show()
```

```python
data = pd.read_csv('data5.csv')

data['Date'] = pd.to_datetime(data['Date']) # convert the column to datetime
data.sort_values('Date', inplace  = True) # pandas using sort_values, not sort

plt.style.use('seaborn')

price_date = data['Date']
price_close = data['Close']


plt.plot_date(price_date, price_close, linestyle = 'solid')

plt.gcf().autofmt_xdate() 


date_formate = mpl_dates.DateFormatter('%b/%d/%Y') 
plt.gca().xaxis.set_major_formatter(date_formate) 

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout() 

plt.show()
```
