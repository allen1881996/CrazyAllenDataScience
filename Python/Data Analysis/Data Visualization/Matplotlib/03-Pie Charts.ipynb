```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")
plt.xkcd()
plt.figure(figsize = (8,8))
slices = [60,40,30,20]
labels = ['Sixty','Fourty','Thirty','Twenty']
explode = (0.1, 0, 0, 0)
colors = ['#6d904f','#e5ae37','#008fd5','#fc4f30']
plt.pie(slices, labels = labels, wedgeprops = {'edgecolor': 'black'}, explode = explode, autopct='%.0f%%', shadow=True, startangle=140, colors = colors)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```
