```python
import random
```

```python
# 0 is inclusive and 1 is non-inclusive
for i in range(10):
    value = random.random()
    formatvalue = 'the {0} number is {1:.2f}'.format(i,value)
    print(formatvalue)
```

```python
# random floating point value between certain range
a = random.uniform(1,10)
print(a)
```

```python
# integer
# 这个是包含2的
a = random.randint(1,2)
print(a)
```

```python
# randomly choice value from a list
first_names = ['John', 'Jane', 'Corey', 'Travis', 'Dave', 'Kurt', 'Neil', 'Sam', 'Steve', 'Tom', 'James', 'Robert', 'Michael', 'Charles', 'Joe', 'Mary', 'Maggie', 'Nicole', 'Patricia', 'Linda', 'Barbara', 'Elizabeth', 'Laura', 'Jennifer', 'Maria']
```

```python
value = random.choice(first_names)
print(value)
```

```python
# mutiple choices from a list
# k is how many times we want to pick a value from the list
values = random.choices(first_names, k = 4)
print(values)
```

```python
# weighted choices
color = ['red','yellow','black']
colors = random.choices(color, weights = [10,2,2], k = 10)
print(colors)
```

```python
# randomly shuffle a list of values
deck = list(range(1, 53))
```

```python
random.shuffle(deck)
print(deck)
```

```python
# 如果希望不重复选择数字
# unique value
hand = random.sample(deck, k=5)
print(hand)
```
