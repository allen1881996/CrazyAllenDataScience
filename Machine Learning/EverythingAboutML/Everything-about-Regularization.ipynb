```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

# Ridge Regression

## Why we need Ridge Regression

Shortcomings of linear regression：
- Predictive ability: Linear regression has low bias (zero bias) but suffers from high variance. So it may be worth sacrificing some bias to achieve a lower variance.

- Interpretative ability: With a large number of predictors, it can be helpful to identify a smaller subset of important variables. Linear regression doesn’t do this.

## What is Ridge Regression

```python
x = np.random.random([2,1])
y = x*2.5 + 1

x2 = np.random.random([10,1]).ravel()
y2 = x2*2.5 + np.random.randn(len(x2))/3 - np.random.random(len(x2))/3 + 1
```

Suppose the red points are our training data, and green points are testing data. 

```python
plt.scatter(x,y,c='r')
plt.scatter(x2,y2,c='g')
```

We fit a line on our two training points, obviously, this linear model has a high variance for our testing data. 

```python
plt.plot(x,y,c='b')
plt.scatter(x2,y2,c='g')
plt.scatter(x,y,c='r')
```

```python
plt.plot(np.linspace(0,1,10),np.linspace(0,1,10)*2.5+1,c='b')
plt.scatter(x2,y2,c='g')
```

Our original linear regression function:

$$ Y = intercept + slope * X $$

$$ minimize: MSE $$

Ridge linear regression function:

$$ Y = intercept + slope * X $$

$$ minimize: MSE + \lambda * slope^2$$

$ \lambda * slope^2 $ adds a penalty to the traditional MSE, and $\lambda$ determines how severe that penalty is.

##  Ridge Regression using sklearn

```python
dataset = pd.read_csv("data/Advertising.csv")
dataset.head()
```

```python
dataset.drop(['Unnamed: 0'], axis=1, inplace=True)
dataset.info()
```

```python
def scatter_plot(feature, target,dataset):
    plt.figure(figsize=(16, 8))
    plt.scatter(
        dataset[feature],
        dataset[target],
        c='black'
    )
    plt.xlabel("Money spent on {} ads ($)".format(feature))
    plt.ylabel("Sales ($k)")
    plt.show()
```

```python
scatter_plot('TV','sales',dataset)
```

```python
scatter_plot('radio','sales',dataset)
```

```python
scatter_plot('newspaper','sales',dataset)
```

```python
dataset.columns
```

```python
from mpl_toolkits.mplot3d import Axes3D

plt.figure(figsize=(20,12))
ax = plt.subplot(111, projection='3d') 
img = ax.scatter(dataset["TV"], dataset["radio"], dataset["newspaper"],c=dataset["sales"],s = 100,cmap='rainbow')  

ax.set_zlabel('TV') 
ax.set_ylabel('radio')
ax.set_xlabel('newspaper')
plt.colorbar(img)
plt.show()
```

```python
X = dataset.drop(['sales'], axis=1)
Y = dataset['sales'].values.reshape(-1,1)

lin_reg = LinearRegression()

MSEs = cross_val_score(lin_reg, X, Y, scoring='neg_mean_squared_error', cv=5)

mean_MSE = np.mean(MSEs)

print(mean_MSE)
```

```python
alpha = [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20]

ridge = Ridge()

parameters = {'alpha': alpha}

ridge_regressor = GridSearchCV(ridge, parameters,scoring='neg_mean_squared_error', cv=5)

ridge_regressor.fit(X, Y)
```

```python
ridge_regressor.best_params_
```

```python
# This is a slight improvement upon the basic multiple linear regression
ridge_regressor.best_score_
```

```python
lasso = Lasso()

parameters = {'alpha': alpha}

lasso_regressor = GridSearchCV(lasso, parameters, scoring='neg_mean_squared_error', cv = 5)

lasso_regressor.fit(X, Y)
```

```python
lasso_regressor.best_params_
```

```python
lasso_regressor.best_score_
```

```python
plt.figure(figsize=(15,10))
plt.plot(lasso_regressor.predict(X),c='r')
plt.plot(ridge_regressor.predict(X),c='g')
plt.plot(Y,c='r')
plt.show()
```
