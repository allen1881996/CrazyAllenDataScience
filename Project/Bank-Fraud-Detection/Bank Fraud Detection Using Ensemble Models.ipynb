## Data Preprocessing 

### Distinguish categorial and continuous variables 

```python
unique_values = {}

for i in dataset_train.columns:
    unique_values[i] = len(set(dataset_train[i]))

unique_values
```

```python
# here, I just assume 12 is a threshold of categorial and continuous variables
for i,j in unique_values.items():
    if j < 12:
        print(set(dataset_train[i]))
```

```python
cat_variables = [i for i in unique_values.keys() if unique_values[i] < 12]
con_variables = [i for i in unique_values.keys() if unique_values[i] > 12]
# Y is the response variable, our target
cat_variables.remove("Y")
print("The continuous variables: ",con_variables)
print("The categorial variables: ",cat_variables)
```

### Basic feature analysis

Check the distribution of features, in order to ensure that our training dataset and testing dataset are coming from the same distribution, which is the assumption of our machine learning algorithm.

```python
count = 1

for i in range(len(cat_variables)):
    fig = plt.figure(figsize = (15,25))
    plt.subplot(len(cat_variables),2,count)
    sns.violinplot(dataset_train[cat_variables[i]],palette="hls")
    plt.title("Train")
    
    plt.subplot(len(cat_variables),2,count+1)
    sns.violinplot(dataset_test[cat_variables[i]],palette="Paired")
    plt.title("Test")
    count += 2
```

```python
count = 1

for i in range(len(con_variables)):
    fig = plt.figure(figsize = (15,25))
    plt.subplot(len(con_variables),2,count)
    sns.boxplot(dataset_train[con_variables[i]],palette="hls")
    plt.title("Train")
    
    plt.subplot(len(con_variables),2,count+1)
    sns.boxplot(dataset_test[con_variables[i]],palette="Paired")
    plt.title("Test")
    count += 2
```

The distributions are similiar, which means we can effectively use our model in testing dataset.

If your data is not non-negative, maybe chi2 is not a good method. Pearson's chi square test (goodness of fit) does not apply to negative values.
```python
dataset_train["Y"].value_counts()
```

```python
weight = dataset_train["Y"].value_counts()[0]/dataset_train["Y"].value_counts()[1]
weight
```

Here we notice the label is imbalanced. It will harm our model if we don't deal it very well! 

Method to deal with imbalanced data:

1. Choose a good metrics: Confusion Matrix, Recall/F1-score; However, precision is a very bad evaluation method for imbalanced data. 
2. Resampling Your Dataset:
  * You can add copies of instances from the under-represented class called over-sampling. It's useful for some insensitive data. We won't fabricate data for sensitive dataset like transcation dataset.
  * use systematic algorithms to generate new instances: Synthetic Minority over-sampling Technique.
  * You can delete instances from the over-represented class, called under-sampling, which is not recommended for small dataset. 
  * imbalanced-learn package: https://github.com/scikit-learn-contrib/imbalanced-learn
3. Penalized Models:
In RandomForestClassifier, set class_weight={0:1,1:4} (the weight of 0 is 1, and the weight of 1 is 4), adjust penalty in LogisticRegression, penalized-SVM....

## Modeling

### Baseline model - random forest

Firstly, use random forest model as my first baseline model, then use this performence to further improve our model.

Why chose random forest model?

- By principle since it randomizes the variable selection during each tree split it's not prone to overfit unlike other models.

- You don't need to pre-precessing data too much.

- I used to read that RF don't need CV because it has includes bootstrap sampling and random variable selection in the process of model fit. But still it is strongly recommended to do CV if you can.

```python
Y.shape, X.shape
```

```python
rfModel = RandomForestClassifier(class_weight = {0:1, 1:weight})
rfModel = rfModel.fit(x_train, y_train)
y_pred = rfModel.predict(x_test)
print(metrics.classification_report(y_test, y_pred))
```

### Basic parameter tuning: grid searching

http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html

```python
def search_model(x_train, y_train, est, param_grid, n_jobs, cv):
    model = GridSearchCV(estimator = est,
                                     param_grid = param_grid,
                                     scoring = 'f1_weighted',
                                     # controls the verbosity: the higher, the more messages
                                     verbose = 10,
                                     # number of jobs to run in parallel.
                                     n_jobs = n_jobs,
                                     # return the average score across folds, weighted by the number of samples in each test set.
                                     iid = True,
                                     # cross-validation generator or an iterable
                                     cv = cv)
    # fit grid search model
    model.fit(x_train, y_train)   
    return model
```

```python
param_grid = {'n_estimators':[100,300,500],
             'criterion':['gini', 'entropy'],
             'class_weight': [{0:1, 1:weight}]}

# a prameter grid was given for our model, and we will iterate over different combination of gini and entropy.
# Find the optimal combination according to F1-score.
rfModel = search_model(X.values, Y.values, RandomForestClassifier(), param_grid, -1, 5)
```

```python
print("Best score: %0.3f" % rfModel.best_score_)
print("Best parameters set:", rfModel.best_params_)
```

### Model Ensemble

[5 Easy questions on Ensemble Modeling everyone should know](https://www.analyticsvidhya.com/blog/2015/09/questions-ensemble-modeling/)

```python
clf1 = RandomForestClassifier(n_estimators = 500, criterion = 'entropy',class_weight = {0:1, 1:weight})
clf2 = KNeighborsClassifier(n_neighbors=5)
clf3 = LogisticRegression (class_weight = {0:1, 1:weight})

# results from your gridsearch
eclf = VotingClassifier(estimators=[('Random_Forest',clf1), ('KNN', clf2),('Logistic', clf3)], voting='hard')
for clf, label in zip([clf1, clf2, clf3, eclf], ['Random_Forest', 'KNN','Logistic', 'Ensemble']):
    scores = cross_val_score(clf, X, Y, cv=3, scoring='f1_weighted')
    print ("Accuracy: %0.2f (+/- %0.2f) [%s]" % (scores.mean(), scores.std(), label))
```
